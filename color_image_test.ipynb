```python
src = "media\\Lenna.png"
```

```python
with Image.open(src) as im:
    ycbcr = np.asanyarray(im.convert('YCbCr'))
    rgb = np.asanyarray(im)
    (h,w) = rgb.shape[0:2] # might be the other way around
    r = rgb[:,:,0]
    g = rgb[:,:,1]
    b = rgb[:,:,2]
    r3 = np.dstack((r,r,r))
    g3 = np.dstack((g,g,g))
    b3 = np.dstack((b,b,b))
    
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1,3,1)
plt.imshow(r3)
plt.title("Red chanel")

plt.subplot(1,3,2)
plt.imshow(g3)
plt.title("Green chanel")

plt.subplot(1,3,3)
plt.imshow(b3)
plt.title("Blue chanel")

plt.show()
```

```python
r_bin = g2b(r,0.85)
g_bin = g2b(g, 0.6)
b_bin = g2b(b, 0.6)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1,3,1)
plt.imshow(r_bin, cmap=plt.cm.gray)
plt.title("Red chanel")

plt.subplot(1,3,2)
plt.imshow(g_bin, cmap=plt.cm.gray)
plt.title("Green chanel")

plt.subplot(1,3,3)
plt.imshow(b_bin, cmap=plt.cm.gray)
plt.title("Blue chanel")

plt.show()
```
